# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=[2,3,6,9]
y=[150,650,420,6]
```

```python
plt.plot(x,y)
```

```python
legend=["s","a","m","d"]
plt.xticks(x,legend)
plt.plot(x,legend)
plt.show(x,legend)
```

```python
plt.bar(x,legend)
```

```python
plt.title("Graph")
plt.ylabel("Yaxis")
plt.xlabel("Xaxis")
```
